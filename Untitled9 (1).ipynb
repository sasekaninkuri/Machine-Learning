```python
features = df.drop(["DECLARED_NET_INCOME","OCCUPATIONAL_STATUS_CODE","ACCOUNT_BALANCE","AMT","EVENT_NUMBER","CHANNEL","ACCOUNT_NUMBER","TRANSACTION_REFERENCE","SEX_CODE","NUMBER_OF_ACCOUNTS","AGE"], axis=1)
features
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features , y,
                                                    test_size=0.5,
                                                    random_state = 42)
x_train, x_test, y_train, y_test
```

```python
y_train = y_train.str.replace(',', '')
y_train = y_train.astype(float)
```

```python
y_test = y_test.replace(',', '', regex=True)
y_test = pd.to_numeric(y_test, errors='coerce')
```

```python
x_train = x_train.replace(',', '', regex=True).astype(float)
```

```python
##RANDOM

from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators = 10, max_depth = 200, random_state = 42)
regressor
```

```python
fitter = regressor.fit(x_train, y_train)
fitter
```

```python
predictions = fitter.predict(x_test)
predictions
```

```python
r2 = r2_score(y_test, predictions)
r2
```

```python
mse_random = mean_squared_error(y_test, predictions)
mse_random
```

```python
rmse_random = root_mean_squared_error(y_test, predictions)
rmse_random
```

```python
mae_random = mean_absolute_error(y_test, predictions)
mae_random
```

```python
##LinearRegression

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
model_train = lr.fit(x_train, y_train)
model_train
```

```python
model_test = lr.predict(x_test)
model_test
```

```python
accuracy = r2_score(y_test , model_test)
accuracy
```

```python
mse_Linear = mean_squared_error(y_test, model_test)
mse_Linear
```

```python
rmse_Linear = root_mean_squared_error(y_test, model_test)
rmse_Linear
```

```python
mae_Linear = mean_absolute_error(y_test, model_test)
mae_Linear
```
